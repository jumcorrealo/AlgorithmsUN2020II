# Bubble Sort

```
from itertools import permutations  
import matplotlib.pyplot as plt


def bubblesort(listx):
  pasos = 0
  ifswap = False
  longitud = len(listx)
  for i in range(len(listx)):
    ifswap = False
    for j in range(longitud-1, i+1, -1):
      pasos += 1
      if (listx[j] < listx[j-1]):
        listx[j], listx[j-1] = listx[j-1], listx[j]
        ifswap = True
        pasos += 1
    if (ifswap == False):
      break;
  return pasos




```

```
fig, axs = plt.subplots(ncols=7, nrows=1, gridspec_kw={'hspace': 0, 'wspace' : 0}, figsize=(25, 5));
fig.suptitle("Iterations",fontsize=20, va='center', y=.97, fontstyle='oblique')


axs[0].set_ylabel('Density')

for j in range(3,10):
  permlist = []
  permutation = permutations(range(1,j+1))
  for i in permutation:
    i = list(i)
    permlist.append(bubblesort(i))
  axs[j-3].hist(permlist, density=True, stacked=True, alpha=0.9);
  axs[j-3].set_title('lista len: '+  str(j))
  
```
